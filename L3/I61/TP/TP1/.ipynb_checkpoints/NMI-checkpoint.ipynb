L'information mutuelle normalisee(NMI) et d'abord une mesure de la quantite d'information partager la version normaliser n'indique juste qu'elle est definie entre 0 et 1, la MI est definie par:

2) Considérons l'image "lena.jpeg" comme une variable aléatoire, calculer et afficher la densité de probabilité associée.
